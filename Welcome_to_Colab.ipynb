```
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

puppy_img = Image.open('sample_data/images/00-puppy.jpg')
puppy_img_np = np.asarray(puppy_img)

puppy_red = puppy_img_np.copy()
# To access the red channel of the picture
puppy_red[:,:,0]
# To set the red channel to 0
puppy_red[:,:,0] = 0
plt.imshow(puppy_red)
```
